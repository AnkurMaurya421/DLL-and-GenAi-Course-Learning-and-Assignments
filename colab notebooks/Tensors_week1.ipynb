Tensors Datatype
...convert

```python
float_tensor=torch.tensor([1,2,3],dtype=torch.float64)
bool_tensor=torch.tensor([0,1,2,3,0,0,0,0,1,2],dtype=torch.bool)
bool_tensor2=torch.tensor([True,False,True],dtype=torch.int32)
print(float_tensor)
print(bool_tensor)
print(bool_tensor2)
print(float_tensor.element_size())
print(bool_tensor.element_size())
print(bool_tensor2.element_size())
```

## Tensors reshaping and indexing

```python
t=torch.arange(1,8)
t,t.shape
```

```python
t_reshaped=t.reshape(1,7) #adds 1 dimension
```

```python
t_reshaped
```

```python
t_reshape_with_memoryshare=t.view(1,7)
t_reshape_with_memoryshare
```

```python
t_reshape_with_memoryshare[:,0]=5
t_reshape_with_memoryshare,t
```

```python
t_stacked=torch.stack([t,t,t],dim=0)
t_stacked
```

```python
t_stacked=torch.stack([t,t,t],dim=1)
t_stacked
```

```python
t_squezed=t_reshaped.squeeze() #removes 1 dimension
t_squezed
```

```python
t_unsquezed=t_squezed.unsqueeze(dim=0)
t_unsquezed
```
